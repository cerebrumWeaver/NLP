```python
# 提取 首页相关热点 字段
def extract_home_fields(data):
    # 报错，请找出原因
    #titles_fixed = data['raw_data']['board'][0]['fixed_items'][0]['title']# TypeError: 'NoneType' object is not subscriptable
    hot_ranking = [(
        item['id'], item['title'], item['title_label_desc'], item['title_label_type']
    )
        for item in data['raw_data']['board'][0]['hot_board_items']
    ]
#     return titles_fixed, hot_ranking
    return '测试', hot_ranking
```

```python
def save_comments(fileName, comments, epoch):
    #fileName = fileName.replace('/', '&') + '.csv'
    if os.path.exists(fileName) and epoch == 0:
                os.remove(fileName)
    with open(fileName, mode='a+', encoding='utf-8', newline='') as f:
                csv_writer = csv.writer(f)
                csv_writer.writerows(comments)
    #print('保存文件：' + fileName)
```

```python
# 通过 相关热点 字段 构建 评论链接
def generate_comment_urls(*args):
#     datetime_current = time.strftime('%Y-%m-%d', time.localtime())
#     file_path = os.path.join('./hot_news', datetime_current)
#     if not os.path.exists(file_path):
#         os.makedirs(file_path)
    start = time.time()
    print(args)
    for home_fields in args:
    #for home_id, home_title, _, _ in home_fields:    
        home_id, home_title, _, _ = home_fields
        epoch = 0
        group_id = '&group_id=%d' % home_id
        home_title = home_title.replace('/', '&') + '.csv'
        # 动态加载评论，当data数据返回长度为0跳出循环
        while(True):
            offset = 'offset=%d' % epoch
            url_comment_epoch = url_header_comments + offset + group_id + url_footer_comments
            
            comments = json.loads(req.request('GET', url=url_comment_epoch, headers=headers).text)
            comments_datas = comments['data']
            if(len(comments_datas) > 0 ):
                comment_fields = [(
                    item['comment']['id'], item['comment']['text'],
                    item['comment']['reply_count'], item['comment']['digg_count'],
                    item['comment']['bury_count'],time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(item['comment']['create_time'])),
                    item['comment']['publish_loc_info'], item['comment']['user_id'], item['comment']['user_name']
                )
                    for item in comments_datas
                ]
                #注意：添加字段时要修改形状大小
                save_comments(os.path.join(file_path, home_title), comment_fields, epoch)
                epoch += 40
            else:
                print(home_title + '====》跳出：' + str(epoch))
                break
    end = time.time()
    print(threading.current_thread().getName() + '共耗时：' + str(end - start))
```
